```python
import pandas as pd
import seaborn as sns
import numpy as np
iris = sns.load_dataset("iris")
iris_x = iris.loc[:,['sepal_length', 'sepal_width',
            'petal_length', 'petal_width']]
import random
random.seed(1)
for col in range(4):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
iris_x.head(10)
```

```python
df_dropna_1 = iris_x.dropna(how='any', axis=0)
display(df_dropna_1.head(3))
```

```python
df_dropna_2= iris_x.dropna(thresh=2, axis=0)
display(df_dropna_2.head(10))
```

```python
df_dropna_3 = iris_x.dropna(subset=['sepal_length','sepal_width' ], how='any', axis=0)  
display(df_dropna_3.head(10))
```

```python
df_fillna_1 = iris_x.fillna(0)
display(df_fillna_1.head(10))
```

```python
df_fillna_2 = iris_x.fillna(round(iris_x.mean(),1))
display(df_fillna_2.head(10))
```

```python
df_fillna_3 = iris_x.fillna(10)
display(df_fillna_3.head(10))
```

```python
df_fillna_4 = iris_x.fillna(method='ffill')
display(df_fillna_4.head(10))
```
