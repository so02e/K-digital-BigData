```python
import requests
import re


r = requests.get('http://unico2013.dothome.co.kr/crawling/exercise_bs.html')
r.encoding = 'utf-8'
# print(r.text)

from bs4 import BeautifulSoup
bs = BeautifulSoup(r.text, 'html.parser')


# 1. <h1> 태그의 콘텐츠
print(bs.h1.string)






# 2. 텍스트 형식으로 내용을 가지고 있는 <a> 태그의 콘텐츠와 href 속성값
atags = bs.find_all('a') # 모든 a태그를 찾음
# result set 객체가 리턴되는데 이는 list객체와 유사하다.



for tag in atags:
    print(tag.string,":",tag['href'])
    
    





# 3. <img> 태그의 src 속성값
print(bs.img['src'])






# 4. 첫 번째 <h2> 태그의 콘텐츠
print(bs.h2.string)

print(bs.find('h2').text)






# 5. <ul> 태그의 자식 태그들 중 style 속성의 값이 green으로 끝나는 태그의 콘텐츠
print(bs.ul.find(style=re.compile('green$')).text) #ul태그 자손에서 찾게 된다.

print(bs.find('li',style=re.compile('green$')).text)








# 6. 두 번째 <h2> 태그의 콘텐츠
print(bs.find_all('h2')[1].text) # resultset 객체는 리스트와 유사하다 !
h2 = bs.find_all('h2')
print(h2[1].text)






# 7. <ul> 태그의 모든 자식 태그들의 콘텐츠 
print(bs.ol.text.strip())





# 8. <table> 태그의 모든 자손 태그들의 콘텐츠 
table = bs.find('table')

print(table.text.strip())






# 9. name이라는 클래스 속성을 갖는 <tr> 태그의 콘텐츠
tr = bs.find('tr', class_="name")
print(tr.text)





# 10. target이라는 아이디 속성을 갖는 <td> 태그의 콘텐츠
td = bs.find('td', id="target")
print(td.text)





```
